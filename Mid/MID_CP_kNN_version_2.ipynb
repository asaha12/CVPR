```python
# Import necessary libraries
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras.datasets import cifar10

# Path to the CIFAR-10 dataset
cifar10_dir = '/content/drive/MyDrive/cifer10/test'

# Function to convert an image to grayscale
def convert_to_grayscale(image):
    # Open the image using Pillow (PIL)
    image = Image.open(image)
    # Convert to grayscale
    grayscale_image = image.convert('L')
    return grayscale_image

# Function to load and preprocess the CIFAR-10 dataset
def load_cifar10_grayscale(cifar10_dir):
    class_names = os.listdir(cifar10_dir)
    grayscale_images = []

    for class_name in class_names:
        class_dir = os.path.join(cifar10_dir, class_name)
        for image_file in os.listdir(class_dir):
            image_path = os.path.join(class_dir, image_file)
            grayscale_image = convert_to_grayscale(image_path)
            grayscale_images.append(np.array(grayscale_image))

    return np.array(grayscale_images)

# Load the CIFAR-10 dataset in grayscale
cifar10_grayscale = load_cifar10_grayscale(cifar10_dir)

# Display some grayscale images as 2D arrays
num_images_to_display = 5
plt.figure(figsize=(15, 6))
for i in range(num_images_to_display):
    plt.subplot(1, num_images_to_display, i + 1)
    plt.imshow(cifar10_grayscale[i], cmap='gray')
    plt.title("Image " + str(i + 1))
    plt.axis('off')

plt.show()
```
